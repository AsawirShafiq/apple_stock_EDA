```python
# Downloaded data for APple stock for 5 years
import yfinance as yf

ticker = "AAPL"

data = yf.download(ticker, start = '2020-01-01', end = "2025-01-01")

data.head()
```

```python
data.columns
```

```python
# We should flatten the multi index columns for easier access to the columns

data.columns = ['_'.join(col).strip('_') for col in data.columns]
data.columns
```

```python
#Checking if any null values are present.
data.isnull().sum()
```

```python
# One important metric in stock prices is the daily change in stock price.

data['Daily_return'] = data['Close_AAPL'].pct_change()
data['Daily_return']
```

```python
# Another metric we can add is the 7 Day moving average.

data['7D-MA'] = data['Close_AAPL'].rolling(window=7).mean()
data['7D-MA']
```

```python
# Saving data to csv file.
data.to_csv('APPL.csv')
```

```python
# Now that we have extensively pruformed EDA on this data. We will move to making a program which can predict future prices of this stovk based on the previous patterns.
#For this we will use ARIMA time series forcasting model.

import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

data.index = pd.to_datetime(data.index) # Ensure date is datetime
data = data.sort_index() #Sort by date

data
```

```python
#Fit an arima model (p=5, d=1, q=0 as an example)
model = ARIMA(data['Close_AAPL'], order=(5,1,0))

arima_result = model.fit()

#forcasting the next 10 days
forecast =arima_result.forecast(steps=10)
forecast
```

```python
#Using this stock prices data. We can make a very simplistic trading strategy Moving Averages.

# Buy signal: When short-term moving average (10 day) crosees above the long-term moving average (50 day).
# Sell Signal: When short-term moving Average crosses below the long-term moving average

data.index = pd.to_datetime(data.index)
data = data.sort_index()

data.head()
```

```python
#Compute short term and long term moving averages

data['Short_MA'] = data["Close_AAPL"].rolling(window=10).mean()
data['Long_MA'] = data["Close_AAPL"].rolling(window=50).mean()

data.tail()
```

```python
#Now we implement the stratgey

data['Signals'] = 0

#Buy signal
data.loc[data['Short_MA'] > data['Long_MA'], 'Signal'] = 1

#Sell signal
data.loc[data['Short_MA'] < data['Long_MA'], 'Signal'] = -1

#Indetify the exact buy/sell points

data['Buy_signal'] = (data['Signal'] == 1) & (data['Signal'].shift(1) != 1 )

data['Sell_signal'] = (data['Signal'] == -1) & (data['Signal'].shift(1) != -1 )


data.tail()
```

```python
#visualizing the buy/ sell signals

plt.plot(data['Close_AAPL'], label ="Close Price", alpha = 0.5)

plt.plot(data['Short_MA'], label='Short MA (10-day)', alpha=0.75)
plt.plot(data['Long_MA'], label='Long MA (10-day)', alpha=0.75)

plt.scatter(data.index[data['Buy_signal']], data['Close_AAPL'][data['Buy_signal']], label = 'Buy Signal', marker ='^', color='green', alpha=1)
plt.scatter(data.index[data['Sell_signal']], data['Close_AAPL'][data['Sell_signal']], label = 'sell Signal', marker ='v', color='red', alpha=1)

plt.title("Apple trading strategy: Moving Avergae crossover stragety")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.show()
```
